```python
def odd_num (numbers):
  num_of_odd=0
  for i in range (numbers):
    if i %2 != 0:
      num_of_odd+=1
  return num_of_odd
```

```python
print (odd_num(15))
```

```python
def american_system (floor):
  if floor <= 0:
    return floor
  elif floor >0 and floor <= 13:
    return floor-1
  else:
    return floor-2
```

```python
print(american_system(1))
print(american_system(0))
print(american_system(5))
print(american_system(15))
print(american_system(-3))
```

```python
def avg(numbers):
  sum=0
  for i in numbers:
    sum+=i
  return sum/len(numbers)
```

```python
print(avg([70,80,77,92,73,66]))
```

```python
def make_negative(number):
  if number <0:
    return number
  else:
    return number-(2*number)
```

```python
print (make_negative(1))
print (make_negative(-5))
print (make_negative(0))
```
