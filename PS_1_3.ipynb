```python
def century (year):
  x=year/100
  if x % 2 == 0:
    print(int(x))
  else:
    print(int(x+1))
```

```python
def divisibility (n, x, y):
  if n*x %2 == 0 and n*y %2 == 0:
    print(f"true becouse {n} is divisible by {x} and {y}")
  else:
    print(f"false becouse {n} is neither divisible by {x} nor {y}")
```

```python
divisibility (3,1,3)
```
