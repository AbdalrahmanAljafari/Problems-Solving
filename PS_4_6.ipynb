```python
def reverse_string(input_text):
    reversed_string = ""

    for i in range(len(input_text) - 1, -1, -1):
        reversed_string += input_text[i]
    return reversed_string
```

```python
print (reverse_string("Abdalrahman"))
```

```python
def mango(quantity, price):
  free_mango=  int(quantity/3)
  if free_mango <= 0:
    print("no mangos free total cost: $",(quantity*price))
  elif free_mango > 0 :
    print (f"You got {free_mango} mangoes for free, total cost: $" ,((quantity-free_mango)*price))
  else:
    print("ivalid enterd")
```

```python
print(mango(5, 3))
```
