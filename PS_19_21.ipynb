```python
def possible_check(pump_away, miles_per_gallon, gallons_left):
  check = gallons_left*miles_per_gallon
  if check >= pump_away:
    return ("true")
  else:
    return ("false")
```

```python
print(possible_check(50,25,2))
print(possible_check(100,50,1))
```
