```python
def minimum(find_min):
    after_min = []
    for i in find_min:
        after_min.append(i)
    return "min= ", min(after_min)

def maximum(find_max):
    after_max = []
    for x in find_max:
        after_max.append(x)
    return "max= ", max(after_max)
```

```python
print(minimum([1,2,3,4]))
print(maximum([1,2,3,4]))
```

```python
def litres(hour):
    quantity_leters=int(hour*0.5)
    return quantity_leters
```

```python
print(f"time = 3 ----> litres = {litres(3)}")
```

```python
def multiplying(numbers):
  result=1
  for i in numbers:
    result*=i
  return result
```

```python
print(multiplying([1,2,3,4]))
```
