```python
def math (opperator, value1, value2):
  if opperator == '+':
    return value1+value2
  elif opperator == '-':
    return value1-value2
  elif opperator == '*':
    return value1*value2
  elif opperator == '/':
    return value1/value2
  else:
    print ("Unknown opperator")
```

```python
print(math('+',4,7))
print(math("-",15,18))
print(math('*',5,5))
print(math('/',49,7))
```

```python
def sheep_count(Exists_or_not):
    count_result = 0
    for i in Exists_or_not:
        if i == 'true' or i == 'True':
            count_result += 1
    return count_result
```

```python
print(sheep_count(['false',"false",'false',
                   'true','true','True',
                   'false','True',"True",
                   "true","false","false"]))
```

```python
def square_sum(values):
  square=0
  sum=0
  for i in values:
    square=i**2
    sum+=square
  return sum
```

```python
print (square_sum([1,2,2]))
```
