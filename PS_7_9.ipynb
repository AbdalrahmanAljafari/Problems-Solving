```python
def quarter(month):
    first_quarter = [1, 2, 3]
    second_quarter = [4, 5, 6]
    third_quarter = [7, 8, 9]
    forth_quarter = [10, 11, 12]

    if month in first_quarter:
        print(f"month {month}, is part of first quarter")
    elif month in second_quarter:
        print(f"month {month}, is part of second quarter")
    elif month in third_quarter:
        print(f"month {month}, is part of third quarter")
    elif month in forth_quarter:
        print(f"month {month}, is part of forth quarter")
    else:
        print("wrong entry")
```

```python
quarter(11)
```

```python
def invert(list_numbers):
    additive_inverse = []
    for element in list_numbers:
        additive_inverse.append(element * -1)
    return additive_inverse
```

```python
invert([1,-2,3,-4,5])
```

```python
def remove(text):
    after_remove = []
    for i in text:
        after_remove.append(i)
    after_remove = "".join(after_remove)
    after_remove = after_remove.replace("!", "")
    return after_remove
```

```python
print(remove("Abdalrahman!!!!!!!!!! Aljafari"))
```
